```python
import pandas as pd
```

```python
# load datasets
df = pd.read_csv("employee_attrition.csv")
```

```python
# Basic overview
df.info()
df.head()
```

```python
# Missing values
df.isna().sum()
```

```python
# Overall attrition count
df['Attrition'].value_counts()
```

```python
# Attrition rate by department
df.groupby('Department')['Attrition'].value_counts(normalize=True).unstack()
```

```python
# Average job satisfaction by attrition
df.groupby('Attrition')['Job_Satisfaction'].mean()
```

```python
# Attrition vs Overtime
pd.crosstab(df['Overtime'], df['Attrition'], normalize='index')
```
