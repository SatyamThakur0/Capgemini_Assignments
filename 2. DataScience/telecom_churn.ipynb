```python
import pandas as pd
```

```python
# load datasets
churn_df = pd.read_csv("telecom_churn.csv")
```

```python
# Basic overview
churn_df.info()
churn_df.head()
```

```python
# Overall churn rate
churn_df['Churn'].value_counts(normalize=True)
```

```python
# Churn by Contract type
pd.crosstab(churn_df['Contract_Type'], churn_df['Churn'], normalize='index')
```

```python
# Average Monthly Charges by churn
churn_df.groupby('Churn')['Monthly_Charges'].mean()
```

```python
# Average tenure by churn
churn_df.groupby('Churn')['Tenure_Months'].mean()
```

```python
# Churn by Internet Service
pd.crosstab(churn_df['Internet_Service'], churn_df['Churn'], normalize='index')
```
