```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'Annual_Income': np.random.uniform(15, 100, 1000),
    'Spending_Score': np.random.uniform(5, 40, 1000)
}
df = pd.DataFrame(data)
df.head()
```

```python
scalar = StandardScaler()
X_scaled = scalar.fit_transform(df)
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
X = [i for i in range(1, 11)]
plt.plot(
    X,
    wcss,
    marker='.',
    color='gray',
    markerfacecolor='red'
    )
for x,y in zip(X, wcss):
    plt.text(
        x, y,
        f"({x},{y:.1f})",
        fontsize=8,
        rotation=45,
        va = 'bottom',
        ha='left'
    )
plt.show()
```

```python
model = KMeans(n_clusters=4, random_state=42)
clusters = model.fit_predict(X_scaled)
df['Cluster'] = clusters

plt.scatter(
    df['Annual_Income'],
    df['Spending_Score'],
    c=df['Cluster'],
    marker='.',
    cmap='viridis'
)

centers = scalar.inverse_transform(kmeans.cluster_centers_)
centers
plt.scatter(centers[:,0], centers[:,1], color='red') 

plt.show()
```

```python
score = silhouette_score(X_scaled, clusters)
print(f"Silhouette Score: {score.__round__(4)}")
```
