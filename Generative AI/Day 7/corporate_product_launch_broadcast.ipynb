```python
import os
import asyncio
from openai import OpenAI

# NVIDIA Build API
client = OpenAI(
    base_url="https://integrate.api.nvidia.com/v1",
    api_key=os.getenv("NVIDIA_API_KEY")
)

MODEL = "meta/llama-3.1-70b-instruct"

# Broadcast message
broadcast_message = """
Product X launch in Q3.
Target market: North America.
Budget: $5M.
"""


# Generic agent function
async def run_agent(role, responsibility):

    prompt = f"""
You are the {role} in a corporate product launch planning team.

Launch Announcement:
{broadcast_message}

Your responsibility:
{responsibility}

Provide a concise action plan.
"""

    response = client.chat.completions.create(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}],
        temperature=0.4,
        max_tokens=200
    )

    return role, response.choices[0].message.content


# Coordinator Agent broadcasting to all departments
async def broadcast():

    tasks = [
        run_agent(
            "Marketing Agent",
            "Design marketing campaigns, choose channels, define product positioning."
        ),
        run_agent(
            "Finance Agent",
            "Allocate the $5M budget and forecast ROI."
        ),
        run_agent(
            "Operations Agent",
            "Plan manufacturing, logistics, and supply chain preparation."
        ),
        run_agent(
            "Legal Agent",
            "Ensure regulatory compliance, contracts, and risk mitigation."
        ),
        run_agent(
            "HR Agent",
            "Plan staffing, hiring needs, and employee training."
        )
    ]

    results = await asyncio.gather(*tasks)
    return results


# Decision Agent integrates responses
def decision_agent(responses):

    compiled = "\n\n".join([f"{role} Response:\n{text}" for role, text in responses])

    prompt = f"""
You are the Decision Agent coordinating the corporate launch plan.

Launch Announcement:
{broadcast_message}

Department Responses:
{compiled}

Combine these into a Final Corporate Product Launch Plan.
"""

    response = client.chat.completions.create(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3,
        max_tokens=400
    )

    return response.choices[0].message.content


# Main workflow
async def main():

    print("📢 Coordinator Agent broadcasting launch announcement...\n")

    responses = await broadcast()

    print("📥 Department Responses:\n")

    for role, text in responses:
        print(f"--- {role} ---")
        print(text)
        print()

    print("🧠 Decision Agent creating final launch plan...\n")

    final_plan = decision_agent(responses)

    print("🚀 Final Corporate Launch Plan:\n")
    print(final_plan)


# Run system
# asyncio.run(main())
await main()
```
