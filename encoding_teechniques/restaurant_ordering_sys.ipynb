```python
# Scenario: Restaurant Ordering System
# A restaurant wants to build a machine learning model to predict order preparation time.
# They collect data about each order, including:
# - Meal Type: Breakfast, Lunch, Dinner
# - Spice Level: Mild, Medium, Hot
# Since models can’t directly work with text labels, the restaurant uses Label Encoding to 
# convert these categories into numbers.
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_excel('restaurant_order_data.xlsx')
```

```python
le_mt = LabelEncoder()
df['meal_type_encoded'] = le_mt.fit_transform(df['Meal Type'])
le_sl = LabelEncoder()
df['spice_level_encoded'] = le_sl.fit_transform(df['Spice Level'])
```

```python
print("Meal Type Mapping: ",dict(zip(le_mt.classes_,
                                le_mt.transform(le_mt.classes_))))
print("Spice Level Mapping: ",dict(zip(le_sl.classes_,
                                le_sl.transform(le_sl.classes_))))

decoded_meal_type = le_mt.inverse_transform([0,1,2])
print(f"Decoded Meal Type: {decoded_meal_type}")
```

```python
decoded_meal_type = le_mt.inverse_transform([0,1,2])
print(f"Decoded Meal Type: {decoded_meal_type}")
decoded_spice_level = le_sl.inverse_transform([0,1,2])
print(f"Decoded Spice Level: {decoded_spice_level}")
```
