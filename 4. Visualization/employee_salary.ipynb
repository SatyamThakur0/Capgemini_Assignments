```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample salary data
np.random.seed(42)
salaries = np.random.normal(loc=50000, scale=10000, size=1000)

# Plot histogram
plt.hist(salaries, bins=30)
plt.xlabel('Salary')
plt.ylabel('Number of Employees')
plt.title('Distribution of Employee Salaries')

# Show plot
plt.show()
```
