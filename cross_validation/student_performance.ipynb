```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('student_exam_dataset.xlsx')
```

```python
X = df.drop(columns=['Student ID',  'Department', 'Exam Score'], axis=0)
y = df['Exam Score']
```

```python
model = LinearRegression()

kf = KFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(model, X, y, cv=kf, scoring='r2')

print('R2 Scores for each fold: ',scores.round(3))
print("mean R2:", scores.mean().round(3))
```

```python
if scores.std() < 0.05:
    print("model  is stable accross folds.")
else:
    print("Model performance varies across folds, investigate further")
```
