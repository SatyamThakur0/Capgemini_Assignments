```python
# Scenario Question

# A biomedical researcher is designing a convolutional neural network (CNN) to analyze microscopic grayscale
# images of blood cells (64×64 pixels). She begins with a single convolutional layer defined as follows:

# Input: 64×64 grayscale images (1 channel)
# Convolutional Layer: 16 filters, each of size 5×5
# Stride: 1
# Padding: Same
# Activation Function: ReLU
# The goal of this layer is to extract local structural features such as cell boundaries and texture variations,
# which will later help in classifying whether the blood cells are healthy or show signs of abnormality.
```

```python
import tensorflow as tf
from tensorflow import keras

# Single Conv2D layer
conv = keras.layers.Conv2D(
    filters     = 16,          # 16 filters -> 16 feature maps
    kernel_size = (5, 5),      # 5x5 filter
    strides     = (1, 1),
    padding     = 'same',      # SAME padding
    activation  = 'relu',
    input_shape = (64, 64, 1)  # 64x64 grayscale image
)

# Build and summarise
model = keras.Sequential([conv])
model.build(input_shape=(None, 64, 64, 1))
model.summary()

# Output shape: (None, 64, 64, 16)
# Trainable params: 416
```
