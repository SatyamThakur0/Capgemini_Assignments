```python
from dotenv import load_dotenv
from langchain_mistralai import ChatMistralAI
from langchain_core.prompts import ChatPromptTemplate
load_dotenv()


system_prompt = """
You are an AI research assistant.
Your job is to analyze academic text and generate concise summaries
that help students quickly understand research articles.
"""

human_prompt = """
Summarize the following research article in 5 clear bullet points.

Article:
{text}
"""

prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    ("human", human_prompt)
])

article = """
Artificial Intelligence is rapidly transforming healthcare.
Machine learning models can detect diseases from medical images,
predict patient outcomes, and assist doctors in diagnosis.
Recent studies show that deep learning models achieve high
accuracy in identifying cancers from radiology scans.
However, ethical concerns remain regarding data privacy,
bias in algorithms, and the need for explainable AI systems.
"""

model = ChatMistralAI(
    model='mistral-medium-latest'
)
# para = input("Give your Paragraph: ")
final_prompt = prompt.invoke(
    {
        'text': article
    }
)
res = model.invoke(final_prompt)
print('\n','='*130)
print('\n',res.content)
print('\n','='*130)
```

```python
from dotenv import load_dotenv
from langchain_mistralai import ChatMistralAI
from langchain_core.prompts import ChatPromptTemplate

load_dotenv()

model = ChatMistralAI(
    model="mistral-medium-latest",
    temperature=0.3
)

assistant_prompt = """
You are an AI Research Assistant.

Analyze the article and return:

Summary:
Key Insights:
Actionable Recommendation:

Article:
{text}
"""

prompt = ChatPromptTemplate.from_messages([
    ("human", assistant_prompt)
])

article = input("Paste research article:\n")

final_prompt = prompt.invoke({"text": article})

response = model.invoke(final_prompt)

print("\nAI Research Assistant Output:\n")
print(response.content)
```
